```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df_purpose = pd.get_dummies(df['purpose'], drop_first=True)
```

```python
df_purpose.head()
```

```python
df.drop('purpose', axis = 1, inplace =True)
```

```python
df = pd.concat([df_purpose,df],axis = 1)
```

```python
df.head()
```

```python
df.corr()
```

```python
df.drop(['home_improvement', 'delinq.2yrs'], axis = 1, inplace = True)
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
scaled_array = ss.fit_transform(df.drop('not.fully.paid', axis = 1))
```

```python
scaled_array
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = pd.DataFrame(scaled_array,columns=df.columns[:-1])
```

```python
y = df['not.fully.paid']
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
ypred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,ypred)
```
