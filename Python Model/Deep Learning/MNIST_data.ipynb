```
import numpy as np
```

```
x=tf.constant(10)
y=tf.constant(20)
```

```
with tf.Session() as sessobj:
  print(sessobj.run(x+y))
  print(sessobj.run(x-y))
  print(sessobj.run(x*y))
  print(sessobj.run(x/y))
```

```
p=tf.placeholder(tf.int32)
q=tf.placeholder(tf.int32)
```

```
myaddfunc=tf.add(p,q)
```

```
with tf.Session() as sess:
  print(f'Addition : {sess.run(myaddfunc,{p:100,q:200})}')
```

```
mymultifunc=tf.multiply(p,q)
```

```
with tf.Session() as sess:
  print(f'Multiplication : {sess.run(mymultifunc,{p:100,q:200})}')
```

```
mysubfunc=tf.subtract(p,q)
mydividefunc=tf.divide(p,q)
```

```
with tf.Session() as sess:
  print(f'Subtraction : {sess.run(mysubfunc,{p:100,q:200})}')
```

```
with tf.Session() as sess:
  print(f'Division : {sess.run(mydividefunc,{p:100,q:200})}')
```

```
arr1=tf.constant(((1,2,3),(4,5,6),(7,8,9)))
arr2=tf.constant(((9,8,7),(6,5,4),(3,2,1)))
```

```
sess=tf.Session()
sess.run(arr1)
```

```
sess.run(tf.matmul(arr1,arr2))
```

```
arr3=tf.constant(np.random.rand(3,2))
arr4=tf.constant(np.random.rand(2,3))
```

```
func=tf.matmul(arr3,arr4)
```

```
with  tf.Session()  as finalsess:
  print(finalsess.run(func))
```
